# Assignment 2: Deck of Cards

```python
import json
from pprint import pprint
import requests

# https://deckofcardsapi.com/

deck_count = 1

# Define the API URL with number of decks to shuffle [4]
url = f"https://deckofcardsapi.com/api/deck/new/shuffle/?deck_count={deck_count}"

# Make the request to the API
response = requests.get(url)

# If the request is successful, load the JSON content
if response.status_code == 200:
    data = response.json()
    print(json.dumps(data, indent=4))  # Print the formatted JSON data
else:
    print(f"Failed to retrieve data. Status code: {response.status_code}")

# Get the deck ID
deck_id = data["deck_id"]
```

```python
# Draw 5 cards

no_of_cards = 5

# Define the API URL to draw cards
url = f"https://deckofcardsapi.com/api/deck/{deck_id}/draw/?count={no_of_cards}"

# Make the request to the API
response = requests.get(url)

# If the request is successful, load the JSON content
if response.status_code == 200:
    data = response.json()
    with open("cards.json", "w") as fp:
        print(json.dumps(data, indent=4))  # Print the formatted JSON data
else:
    print(f"Failed to retrieve data. Status code: {response.status_code}")
```

```python
card_image_url = []
card_suit = []
card_value = []
for card in data["cards"]:
    print(f"{card['value']} of {card['suit']}")
    card_image_url.append(card['image'])
    card_suit.append(card['suit'])
    card_value.append(card['value'])

```

```python
suits = ['SPADES', 'HEARTS', 'DIAMONDS', 'CLUBS']

# Create a dictionary to store number of cards of each suit
counts = {suit: 0 for suit in suits}

for suit in card_suit:
    counts[suit] += 1

print(counts)
  
```

```python
# Evaluate the hand - basic!
hand = max(counts.values())  

if hand > 5:
  print("Wow - a flush")
  
elif hand == 4:  
  print("Pretty good - four of a kind")
  
elif hand == 3:
  print("A triple - not bad")
  
elif hand == 2:
  print("Only a pair - bluff!")
```
